```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
df = df[0]
df.head()
```

```python
#cleanup column names
df.rename(columns={0:'Postcode', 1:'Borough', 2:'Neighborhood'}, inplace=True)

#delete first row in data
df.drop(df.index[[0]], inplace=True)

#exclude rows where Borough is "Not assigned"
df = df[df.Borough != 'Not assigned']

#assign missing neighborhoods with Borough's name
df.Neighborhood.replace("Not assigned",df.Borough,inplace=True)

#combine neighborhoo data having same postal codes
df = df.groupby('Postcode')['Neighborhood'].apply(', '.join).reset_index()

df
```
